We can see that columns normalized-losses, price, horsepower, peak-rpm and stroke are numerical but still datatypes of them are showing object because these columns contains "?" character in them, we need to replace "?" 

```python
#writing function to replace "?"
def replacer(column_name, symbol):
    mean = df[df[column_name]!=symbol][column_name].astype(float).mean()
    data[column_name] = data[column_name].replace(symbol,mean)
    data[column_name] = data[column_name].astype(float)
    
```

```python
replacer("normalized-losses","?")
replacer("price","?")
replacer("horsepower","?")
replacer("peak-rpm","?")
replacer("stroke","?")
replacer("bore","?")
```
